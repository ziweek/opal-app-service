```python
import pandas as pd


def jaccard_similarity(list1, list2):
    s1 = set(list1)
    s2 = set(list2)
    return float(len(s1.intersection(s2)) / len(s1.union(s2)))


def get_top_list_by_sim(df, column_name):
    df_target = df[[column_name, "j_sim"]]
    df_target_groupby = df_target.groupby([column_name]).mean()
    df_target_groupby_sorted = df_target_groupby.sort_values("j_sim", ascending=False)
    return df_target_groupby_sorted


df_data = pd.read_excel("./분석파일 (2).xlsx")


target_features = ["연령", "응답자 성별"]
li_data = [list(e[1][target_features]) for e in df_data.iterrows()]


inp = int(input())
container = []
for i, e in enumerate(li_data):
    args = li_data[inp]
    sim = jaccard_similarity(args, e)
    context = sim
    container.append(context)


df_data["j_sim"] = container
df_sorted = df_data.sort_values("j_sim", ascending=False)
df_sorted_dropped = df_sorted.drop(inp, axis=0)


get_top_list_by_sim(df_sorted_dropped, "가장 만족스런 여가활동 (1순위)")
```
